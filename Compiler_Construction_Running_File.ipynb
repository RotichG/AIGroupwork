### **Flex**

Flex is (fast lexical analyzer) used to generate a scanner specified by regular expressions and C code for actions to implement it. The scanner reads an input sequence that
matches a pattern, it executes the action to process it allowing the tool to be used in many different contexts due to generalization.
It is faster than Lex and often utilized with Bison.

Initiated by creating a .l file thats passed through the flex to form a lex.yy.c file compiled to create output that is used to create sequences of tokens from an input stream.

### **BISON**
Bison is a general-purpose parser generator that converts a grammar description in form of Bison Grammar Files for a LALR context free grammar into a C program to parse that program.

LALR Parser is lookahead LR parser. It is  the most powerful parser which can handle large classes of grammar.

The parser obtains a sequence of tokens from the lexical analyzer and recognizes its structure in the form of a parse tree.

The parse tree expresses the hierarchical structure of input data and maps grammar symbols to data elements.

The tree nodes represent the symbols of grammar and tree edges represent derivation steps.

Parsing can be top-down or bottom-up.

Top-down starts on the left-hand side while bottom-up starts on the right-hand side.

Bison is a bottom-up parser, and it tries by shifts and reductions to reduce the input to a single grouping whose symbol is the input’s start symbol.

The format of Bison code is as follows:

      C Declarations
    %}
      Bison Declarations
    %%
      Grammar Rules
    %%
      Additional C Code

##Predictive Parsing

Predictive parsing is a simple form of recursive descent parsing in which the lookahead symbol unambiguously determines the non-terminal symbol used to replace the input string.

##LL(1) Parsing

LL(1) parsing is a top-down parsing that uses one lookahead symbol to parse an input string and requires no backtracking. This is achieved through left to right scanning using the left-most derivation.

##First() and Follow()

First() represents the beginning set of characters of a string that can be obtained by applying a production rule. Follow() assists the compiler to figure out the set of chatacters that should come after applying the first() to an input string. Together they allow the parser to easily derive strings without backtracking and reduced ambiguity, as the parser is aware of what step to take next by looking at the parser table constructed from the obtained values of First() and Follow().

##Unique Attributes of Top Down Parsing



1.   It involves matching a production rule to the input string with the aim of acquiring the input string from the production rules.
2.   The depth of the tree is considered first before the breadth.
3.   Can invlove backtracking.
4.   Uses the Left Most Derivation to parse.
5.   Construction of the parse tree begins from the root to the leaves.

## Front End Compiler

The front end compiler has 2 stages:
1. Scanning
2. Parsing

The front end compiler recieved the sorce code as input, which goes through scanning and produces tokens.
The tokens are then parsed and the output will be the intermediate representation which is then used as input to the back end compiler or middle end in the case for a three pass compiler

Scanning and Parsing can be further be broken down into the 4 stages shown in the diagram

### Lexical Analysis

Input - source code
Output - tokens
At this stage, the distinct characters in the code are identified and a pair of role and words are produced as output.

Example;x+2*y
1. Tokens produced:
2. id : x,y
3. number:2
4. symbols: +, *

### Syntax Analysis

Input - Tokens from lexical analyzer
 
Parsing begins at this stage and the parser must distinguish between valid and invalid sequences of tokens based on the structure of the context free grammar.

Example of a CFG:
1. expr ? expr op expr
2.       | num
3.       | id
4. op   ?  +
5.       | -
6.       | *
7.       | /

using the tokens and right-most derivation:
1. expr ? expr op expr
2.        expr op id 'y'
3.        exp  *  id 'y'
4. exp op exp  *  id 'y'
5. exp op num'2'  *  id 'y'
6. exp +  num'2'  *  id 'y'
7. id'x' +  num'2'  *  id 'y'



### Semantic Analysis

The Context free grammar used in syntax analysis is then put into context. 
Checks such as if x and y have been defined.

### Intermediate Code Representation

The parser produces the intermediate code representation after parsing the code.

Example using three adress code: x+2*y

1. t1 = 2*y
2. t2 = x
3. t3 = t2 + t1
